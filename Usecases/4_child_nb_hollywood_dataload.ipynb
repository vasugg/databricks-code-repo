## Loading Hollywood API data into a new `table`


```python

%sql
CREATE VOLUME IF NOT EXISTS workspace.default.HollywoodStories;
```

```python
import requests
response = requests.get("https://public.tableau.com/app/sample-data/HollywoodsMostProfitableStories.csv")
dbutils.fs.put("/Volumes/workspace/default/HollywoodStories/HollywoodsMostProfitableStories.csv", response.text, overwrite=True)
```

```python
%python
df = spark.read.csv(
    "/Volumes/workspace/default/HollywoodStories/HollywoodsMostProfitableStories.csv",
    header=True,
    inferSchema=True,
    samplingRatio=0.5
)

# Replace spaces and parentheses in column names with underscores
df = df.toDF(*[
    col.replace(" ", "_")
       .replace("(", "")
       .replace(")", "")
       .replace("'", "")
       .replace("-", "_")
       .replace(".", "_")
       for col in df.columns
])

df.write.saveAsTable("HollywoodsMostProfitableStories")
```
