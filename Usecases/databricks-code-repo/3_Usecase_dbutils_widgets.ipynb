```python
# MAGIC ###Widgets utility used for adding the components/widgets into our notebook for creating
# MAGIC dynamic/parameterized approaches
#create widget text box to enter the table name
dbutils.widgets.text("tablename","wimbledons_champions","Enter the table name to query!")
#get the table name using text name attribute
tblname=dbutils.widgets.get("tablename")
#spark sql to select all records in the table
sparkdf1=spark.sql(f"select * from workspace.default.{tblname} limit 10")
#display the dataframe
display(sparkdf1)
```

```python

#dropdown list to select Genre
#sparkdf2=spark.sql(f"select distinct genre from workspace.default.hollywood")
#display(sparkdf2)
dbutils.widgets.dropdown("genre","Comedy",["Comedy","Drama","Romance","Action","Animation","Fantasy"],"Select genre")
genre_select=dbutils.widgets.get("genre")
sparkdf2=spark.sql(f"select * from workspace.default.hollywoodsmostprofitablestories where genre='{genre_select}'")
display(sparkdf2)
print(f"Number of films in {genre_select} genre")
display(sparkdf2.count())
```

```python
#Multiselect
dbutils.widgets.multiselect("genre","Comedy",["Comedy","Drama","Romance","Action","Animation","Fantasy"],"Select multi genre")
genre_list=dbutils.widgets.get("genre").split(",")
for i in genre_list:
    print(i)
```

```python
#Combobox
dbutils.widgets.combobox("batch","we47",["we44","we45","we46","we47"],"Select batch ID")
genre_combo=dbutils.widgets.get("batch")
print(genre_combo)
```
