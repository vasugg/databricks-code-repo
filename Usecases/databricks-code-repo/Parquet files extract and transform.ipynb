```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Protob Conversion to Parquet") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()

# read csv
df = spark.read.csv("/Volumes/workspace/default/HollywoodStories/HollywoodsMostProfitableStories.csv")

# Displays the content of the DataFrame to stdout
df.show()

df.write.parquet("/Volumes/workspace/default/HollywoodStories/HollywoodsMostProfitableStories.parquet")
```
