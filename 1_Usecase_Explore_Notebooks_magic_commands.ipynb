###### Task7: Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command %python

```python
%python

import pandas as pd
from pyspark.sql import SparkSession

df1 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/firstcatalog/default/usage_metrics/mobile_os_usage.csv")

for col in df1.columns:
    new_col = col.replace(" ", "_")\
                 .replace(",", "")\
                 .replace(";", "")\
                 .replace("{", "")\
                 .replace("}", "")\
                 .replace("(", "")\
                 .replace(")", "")\
                 .replace("\n", "")\
                 .replace("\t", "")\
                 .replace("=", "")
    df1 = df1.withColumnRenamed(col, new_col)
# Save as table
df1.write.option("inferSchema", "true").saveAsTable("firstcatalog.default.mobile_os_usage")

# Write the PySpark DataFrame to the specified table
# The 'overwrite' mode will replace the table and its data if it already exists
# spark_df.write.mode("overwrite").saveAsTable("firstcatalog.default.mobile_os_usage")
```

##### Task9: Create a python function to convert the given input to upper case

```python
def convert_to_uppercase(input):

  return input.upper()

# Example usage:
first_string = "Hello World!"
uppercase_string = convert_to_uppercase(first_string)
print(f"Original string: {first_string}")
print(f"Uppercase string: {uppercase_string}")

second_string = "python programming"
uppercase_another = convert_to_uppercase(second_string)
print(f"Original string: {second_string}")
print(f"Uppercase string: {uppercase_another}")
```
