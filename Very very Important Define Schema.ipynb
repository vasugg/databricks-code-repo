```python
#Very Very IMPORTANT: 2. Using structure type to define schema.
#For the given data "id, fname, lname, age, prof" I am going to use StructType(StructField("id", IntegerType(), True), StructField("fname", StringType(), True), StructField("lname", StringType(), True), StructField("age", IntegerType(), True), StructField("prof", StringType(), True))

#StructType is a class in PySpark that represents a schema for a DataFrame. It is used to define the structure of a DataFrame, including the names and types of its columns. The StructType class is a subclass of the DataType class and is used to create a schema for a DataFrame that contains multiple columns.

#The StructField class is a class in PySpark that represents a field in a schema. It is used to define the name and type of a column in a DataFrame. The StructField class is a subclass of the DataType class and is used to create a schema for a DataFrame that contains multiple columns.

#The DataType class is a class in PySpark that represents the type of a column in a DataFrame. It is used to define the type of a column in a DataFrame. The DataType class is a base class for all data types in PySpark and is used to create a schema for a DataFrame that contains multiple columns.

#StructType -- 1 time
#StructField -- Total number of field items(here its 5)
#IntegerType -- Total number of integer items(here its 2)
#StringType -- Total number of string items(here its 3)

from pyspark.sql.types import StructType, StructField, IntegerType, StringType
custom_schema = StructType([StructField("id", IntegerType(), True),
    StructField("fname", StringType(), True),
    StructField("lname", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("prof", StringType(), True)])
    
df = spark.read.csv("/Volumes/workspace/default/volumewe47/raw/customer.csv", custom_schema, header=True)
display(df)
```
